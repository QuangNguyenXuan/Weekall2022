```python
# dot product

import numpy as np

x = np.array([4, 2, 1, 2])
y = np.array([1, 2, 2, 0])

dproduct = np.dot(x, y)
print(dproduct)
```

```python
# norm

import numpy as np

x = np.array([4, 2, 1, 2])
norm = np.linalg.norm(x)
print(norm)
```

```python
# cosine similarity

import numpy as np
from numpy import dot
from numpy.linalg import norm

x = np.array([4, 2, 1, 2])
y = np.array([1, 2, 2, 0])

cos_sim = dot(x, y) / (norm(x)*norm(y))
print(cos_sim)
```

```python
u = 2*x
v = 3*y

cos_sim = dot(u, v) / (norm(u)*norm(v))
print(cos_sim)
```
