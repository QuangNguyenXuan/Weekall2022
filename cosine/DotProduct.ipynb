## Vector and matrix operations

```python
import numpy as np

matrix = np.random.randint(0, 9, (3, 3))
print(matrix)
```

```python
vector = np.random.randint(0, 9, (3,))
print(vector)
```

```python
print(matrix@vector)
```

```python
print(vector@matrix)
```

```python
# better

matrix = np.random.randint(0, 9, (3, 3))
print(matrix)

vector = np.random.randint(0, 9, (3, 1))
print(vector)
```

```python
print(matrix@vector)
print(vector@matrix)
```

## Convert color images to grayscale images

```python
# application

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('aloe1.png')

# show image
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
plt.axis('off')
plt.show()
```

```python
im_rgb.shape
```

```python
# prepare vector
vector = np.array([0.213, 0.715, 0.072])
print(vector.shape)
```

```python
img_gray = im_rgb@vector
print(img_gray.shape)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('aloe_gray.png', img_gray)
```
